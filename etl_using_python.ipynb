```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
from sqlalchemy import create_engine
import io
import boto3
#boto3 is the package for uploading your DataFrame to AWS S3(boto3 library, which is the AWS SDK(Software Development Kit) for Python. 
#It allows you to interact with various AWS services, including S3, EC2, Lambda, etc.)

from azure.storage.blob import BlobServiceClient, BlobClient, ContainerClient
#The azure-storage-blob package allows Python applications to easily interact with Azure Blob Storage for uploading, downloading, and managing files in the cloud.
```

```python
#SpotifyClientCredentials is a method in the Spotify Web API for authenticating an application using client ID and client secret.(permission to extract data from spotify)

client_credentials_manager = SpotifyClientCredentials(client_id="client_id", client_secret="client_secret")
```

```python
#Spotify - It is function which is used to extracting the data from spotify by creating an object(used for authorization)

sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
playlist_link = "https://open.spotify.com/playlist/37i9dQZF1DWTt3gMo0DLxA"
```

```python
playlist_uri=playlist_link.split("/")[-1]
```

```python
data = sp.playlist_tracks(playlist_uri)
```

```python
album_list=[]
for row in data["items"]:
    album_id = row["track"]["album"]["id"]
    album_name = row["track"]["album"]["name"]
    release_date = row["track"]["album"]["release_date"]
    url = row["track"]["album"]["external_urls"]["spotify"]
    total_tracks = row["track"]["album"]["total_tracks"]
    album_dict = {"ID":album_id,"Name":album_name,"Release_date":release_date,"URL":url,"Total_tracks":total_tracks}
    album_list.append(album_dict)
```

```python
album_list #check whether the data is arranged properly or not
```

```python
artist_list=[]
for row in data["items"]:
    for key,value in row.items():
        if key=="track":
            for artist in value["artists"]:
                artist_dict={"Id":artist["id"],"Name":artist['name'],"href":artist["href"]}
                artist_list.append(artist_dict)
```

```python
artist_list
```

```python
song_list = []
for row in data["items"]:
    song_id = row["track"]["id"]
    song_name = row["track"]["name"]
    duration = row["track"]["duration_ms"]
    popularity = row["track"]["popularity"]
    track_number = row["track"]["track_number"]
    song_add_date = row["added_at"]
    album_id = row["track"]["album"]["id"]
    artist_id = row["track"]["album"]['artists'][0]["id"]
    song_dict = {"Id":song_id,"Name":song_name,"Duration":duration,"Popularity":popularity,"Track_number":track_number,
              "Add_date":song_add_date,"Album_id":album_id,"Artist_id":artist_id}
    song_list.append(song_dict)
```

```python
song_list
```

```python
# Converting the list data structure to Dataframes

album_df = pd.DataFrame.from_dict(album_list)

artist_df = pd.DataFrame.from_dict(artist_list)

song_df = pd.DataFrame.from_dict(song_list)
```

```python
# Uploading the data into Amazon S3

AWS_ACCESS_KEY = "aws_access_key"

AWS_SECRET_KEY = "aws_secret_key"

# If you don't have a access_key and secret key 

# Visit AWS console --> Login with your credentials --> Go-to-IAM -->

# Converting the pandas dataframe into csv files
# album key is nothing but file name


s3 = boto3.client("s3",aws_access_key_id = AWS_ACCESS_KEY,aws_secret_access_key = AWS_SECRET_KEY)

#The s3 is a object that can be used to perform operations like uploading files, listing buckets, and managing objects in Amazon S3.
# s3 - This is an instance of the boto3.client('s3'), which you use to interact with the Amazon S3 service.
#put_object(): This method is used to upload or store data in an S3 bucket

bucket_name="spotifyv"


filename_1 = "album.csv"
album_buffer = io.StringIO()
album_df.to_csv(album_buffer,index=False)
album_content = album_buffer.getvalue()


filename_2 = "artist.csv"
artist_buffer = io.StringIO()
artist_df.to_csv(artist_buffer,index=False)
artist_content = artist_buffer.getvalue()


filename_3 = "song.csv"
song_buffer = io.StringIO()
song_df.to_csv(song_buffer,index=False)
song_content = song_buffer.getvalue()
```

```python
s3.put_object(Bucket=bucket_name, Key=filename_1, Body=album_content)

s3.put_object(Bucket=bucket_name, Key=filename_2, Body=artist_content)

s3.put_object(Bucket=bucket_name, Key=filename_3, Body=song_content)
```

```python
# Uploading the files into the azure storage account

# Coverting the dataframe into csv file

album_data_buffer = io.StringIO()
album_df.to_csv(album_data_buffer,index=False)

artist_data_buffer=io.StringIO()
artist_df.to_csv(artist_data_buffer,index=False)

song_data_buffer=io.StringIO()
song_df.to_csv(song_data_buffer,index=False)

# StringIO() - It is a module in Python that provides an in-memory stream for text I/O. 
#It allows you to work with strings as if they were files. 
#This can be useful for various purposes, such as when you need to process or manipulate string data using file-like operations.

# account_name means storage account name
account_name="storage_account_name"

# you will find the account_key as Go-to-stroage account --> Network and security --> Access key --> Copy key
account_key="account key"

#container name means your folder name 
container_name="container name" 

# Create a BlobServiceClient object
blob_service_client = BlobServiceClient(account_url=f"https://{account_name}.blob.core.windows.net", credential=account_key)

#BlobServiceClient is used to interact with Azure Blob Storage, allowing you to manage containers and blobs within your storage account.
#account_url - This is the URL of your Azure Storage account, constructed using the account_name.
#credential - This is where you provide the account_key to authenticate your access.
#these steps allows you to interact with the Azure Blob Storage account for tasks like uploading or downloading files.

#Defining the filenames
filename_4= 'album.csv'
filename_5="artist.csv"
filename_6="song.csv"


# Creating the container client
container_client = blob_service_client.get_container_client(container_name)

# Uploading the album_data into Azure blob storage
blob_client = blob_service_client.get_blob_client(container=container_name,blob=filename_4)
blob_client.upload_blob(album_data_buffer.getvalue(),overwrite=True)

# Uploading the artist_data into Azure blob storage
blob_client=blob_service_client.get_blob_client(container=container_name,blob=filename_5)
blob_client.upload_blob(artist_data_buffer.getvalue(),overwrite=True)

# Uploading the song_data into Azure blob storage
blob_client=blob_service_client.get_blob_client(container=container_name,blob=filename_6)
blob_client.upload_blob(song_data_buffer.getvalue(),overwrite=True)
```
